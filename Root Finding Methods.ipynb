### Bisection Method (No of iterations)

```python
from math import *

# defines expression who's root we must find 
def f(x):
#     return math.exp(x) + 2 * x - 7
#     return x**4 + 2*x**3 - x - 1
#     return x*sin(x) - 1
#     return x**3 - 7*x**2 + 14*x - 6
#     return x*e**x - 1
#     return e**(-x) * ((3.2)*sin(x) - (0.5)*cos(x))
    return x**2 - 3

# bisection function returns root of the expression f(x) 
def bisection(firstApprox, secondApprox, n):
    counter = 0
    print("No    a     b     Xn     f(Xn)")
    
    while n > 0:
        counter += 1
        x = (firstApprox + secondApprox) / 2
        print("%2d %0.4f %0.4f %0.4f %0.4f" % (counter - 1, firstApprox, secondApprox, x, f(x)))
        
        if f(x) < 0:
            firstApprox = x
        elif f(x) > 0:
            secondApprox = x
        else:
            break
        n -= 1
    return x

# taking input form user 
a = float(input("Enter first approximation : "))
b = float(input("Enter second approximation : "))
n = int(input("Enter no of iterations: "))

# checking if the roots are bracketed 
print("f(a) = %f" % (f(a)))
print("f(b) = %f" % (f(b)))

if f(a) * f(b) > 0:
    print("Given approximations do not bracket the roots. Try again.")
else:
    print("Root: " + str(bisection(a, b, n)))
```

### Bisection Method (Permissible error)

```python
from math import *

# defines expression who's root we must find 
def f(x):
#     return math.exp(x) + 2 * x - 7
#     return x**4 + 2*x**3 - x - 1
#     return x*sin(x) - 1
#     return x**3 - 7*x**2 + 14*x - 6
    return x*e**x - 1

# bisection function returns root of the expression f(x) 
def bisection(firstApprox, secondApprox, error):
    x = 0 
    counter = 0
    print("No    a     b     Xn     f(Xn)")

    while True:
        counter += 1
        g = f(x)
        x = (firstApprox + secondApprox) / 2
        print("%2d %0.4f %0.4f %0.4f %0.4f" % (counter - 1, firstApprox, secondApprox, x, f(x)))

        if abs(g - f(x)) <= error:
            break

        if f(x) < 0:
            firstApprox = x
        else:
            secondApprox = x

    print("\nNo of iterations: %d" % (counter))
    return x

# taking input form user 
a = float(input("Enter first approximation : "))
b = float(input("Enter second approximation : "))
err = float(input("Enter permissible error: "))

# checking if the roots are bracketed
print("f(a) = %f" % (f(a)))
print("f(b) = %f" % (f(b)))

if f(a) * f(b) > 0:
    print("Given approximations do not bracket the roots. Try again.")
else:
    print("Root: " + str(bisection(a, b, err)))
```

### Newton Raphson Method 

```python
from sympy import *
import math

x = Symbol('x')

# defines expression who's root we must find 
# f = E ** x + 2 * x - 7
# f = x**3 - 5*x + 3
# f = x**2 - 28
f = x**3 - 7*x**2 + 14*x - 6

# fPrime is diffrential of f 
fPrime = f.diff(x)

f =  lambdify(x, f)
fPrime = lambdify(x, fPrime)

# newtonRaphson function returns root of the expression f(x)
def newtonRaphson(x, error):
    counter = 0
    print("No    x     f(x)     f'(x)")

    while True:
        counter += 1
        y = x - f(x) / fPrime(x)
        print("%2d %0.4f %0.4f %0.4f %0.4f" % (counter - 1, x, f(x), fPrime(x), y))
        
        if abs(y - x) <= error:
            print("\nNo of iterations: %d" % (counter))
            return y
        
        x = y

# taking input form user
a = float(input("Enter approximate root: "))
error = float(input("Enter permissible error: "))

print("Root: " + str(newtonRaphson(a, error)))
```

### Secant Method

```python
from math import * 

# defines expression who's root we must find 
def f(x):
#     return math.exp(x) + 2 * x - 7
#     return cos(x) + 2 * sin(x) + x**2
#     return x**4 - x -10
    return x * e**x - 1 

# secant function returns root of the expression f(x)
def secant(x0, x1, error):
    counter = 0
    print("No    Xn")
    
    while True:
        counter += 1
        print("%2d %0.6f" % (counter - 1, x0))
        
        x2 = (x0 * f(x1) - x1 * f(x0)) / (f(x1) - f(x0))
        
        if abs(x2 - x1) <= error:
            print("%2d %0.6f" % (counter, x1))
            print("%2d %0.6f" % (counter + 1, x2))
            print("\nNo of iterations: %d" % (counter))
            return x2
        
        x0 = x1
        x1 = x2

# taking input form user
x0 = float(input("Enter first approximation : "))
x1 = float(input("Enter second approximation : "))
error = float(input("Enter permissible error: "))

print("Root: " + str(secant(x0, x1, error)))
```

### Fixed Point Iteration Method

```python
import math 

# defines expression who's root we must find 

def f(x):
#     return math.exp(x) + 2 * x - 7
#     return  math.log(7 - 2 * x) 
#     return 10 / (x**3 - 1)
#     return (x+10)**(1/4)
#     return ((x + 10)**(1/2)) / x
#     return cos(x) / e**x
#     return 1 / e**x
    return log(1 / x)

# fixedPointIteration function returns root of the expression f(x)
def fixedPointIteration(x0, error):
    counter = 0
    print("No    Xn")
    
    while True:
        counter += 1
        print("%2d %0.6f" % (counter - 1, x0))
        x = f(x0)
        
        if abs(x - x0) <= error:
            print("%2d %0.7f" % (counter, x))
            print("\nNo of iterations: %d" % (counter))
            return x;
        
        x0 = x

# taking input form user     
a = float(input("Enter approximate root: "))
e = float(input("Enter permissible error: "))

print("Root: " + str(fixedPointIteration(a, e)))
```
